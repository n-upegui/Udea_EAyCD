```python
class Cliente:
  """Esta estructura contiene los datos de un cliente"""

  def __init__(self, id, nombre, direccion):
    """Inicializa el cliente

    Keyword arguments:
    id -- identificación del cliente
    nombre -- nombre del cliente
    direccion -- residencia del cliente
    """
    self.id = id
    self.nombre = nombre
    self.direccion = direccion

class Producto:
  """Esta estructura contiene los datos de un producto"""

  def __init__(self, id, nombre, valor):
    """Inicializa el producto

    Keyword arguments:
    id -- identificación del producto
    nombre -- nombre del producto
    valor -- costo del producto
    """
    self.id = id
    self.nombre = nombre
    self.valor = valor

class Pedido:
  """Clase donde se guardan los productos que un cliente va a pedir y que permite
  visualizar el total del pedido
  """

  def __init__(self, cliente):
    """Inicializa el pedido de un cliente

    Keyword arguments:
    cliente -- cliente que realiza el pedido
    """
    self.cliente = cliente
    self.productos = []

  def adicionar_producto(self, producto):
    """Añade un producto al pedido

    Keyword arguments:
    producto -- producto que se va a añadir
    """
    self.productos.append(producto)

  def calcular_valor_total(self):
    """Itera sobre los productos y retorna el valor total"""
    valor = 0
    for producto in self.productos:
      valor += producto.valor
    return valor

  def revisar(self):
    """Imprime la información del pedido"""
    print(f'Cliente: {self.cliente.nombre}')
    for producto in self.productos:
      print(f' {producto.nombre} {producto.valor}')
    print(f' Total: {self.calcular_valor_total()}')
    print(f' Llevar a : {self.cliente.direccion}')
    print()


class PedidoDescuento(Pedido):
  """Clase hija de pedido a la cual se le pueden aplicar descuentos"""

  def __init__(self, cliente, descuento):
    """Inicializa el pedido con descuento para un cliente

    Keyword arguments:
    cliente -- cliente que realiza el pedido
    descuento -- porcentaje que se va a rebajar al total del pedido
    """
    super().__init__(cliente)
    self.descuento = descuento

  def calcular_valor_total(self):
    """Retorna el valor total del pedido despues de hacerle el descuento"""
    return int(super().calcular_valor_total()*(1-self.descuento/100))

  def revisar(self):
    """Imprime la información del pedido"""
    print(f'Cliente: {self.cliente.nombre}')
    for producto in self.productos:
      print(f' {producto.nombre} {producto.valor}')
    print(f' Descuento: {self.descuento}%')
    print(f' Total: {self.calcular_valor_total()}')
    print(f' Llevar a : {self.cliente.direccion}')
    print()

class Inventario:
  """Estructura que permite almacenar los productos registrados por una tienda"""

  def __init__(self):
    """Inicializa el inventario"""
    self.productos = {}

  def adicionar_producto(self, producto):
    """Añade un producto al inventario

    Keyword arguments:
    producto -- producto que se va a añadir
    """
    self.productos[producto.id] = {
        'producto': producto
    }

  def producto_registrado(self, producto):
    """Revisa que el producto exista en el invetario y retorna un booleano

    Keyword arguments:
    producto -- producto que se va a añadir
    """
    return self.productos.get(producto.id)

  def revisar(self):
    """Imprime los productos que hay en el inventario"""
    for id in self.productos:
      print(f' {self.productos[id]["producto"].nombre}')

class TiendaRopa:
  """Esta clase tiene implementa las interfaces para automatizar
  el trabajo de administración y ventas por pedido de una tienda de ropa
  """

  def __init__(self, nombre):
    """Añade un producto al inventario

    Keyword arguments:
    nombre -- nombre de la tienda
    """
    self.nombre = nombre
    self.inventario = Inventario()
    self.historial_pedidos = []
    self.clientes_fieles = []
    self.descuento = 20

  def adicionar_producto(self, producto):
    """Añade un producto al inventario de la tienda

    Keyword arguments:
    producto -- producto que se va a añadir
    """
    self.inventario.adicionar_producto(producto)

  def revisar_inventario(self):
    """Imprime el inventario de la tienda"""
    print(f'Inventario de la tienda {self.nombre}')
    self.inventario.revisar()
    print()

  def registrar_cliente_fiel(self, cliente):
    """Añade un cliente a la lista de clientes que reciben descuentos en su
    pedido

    Keyword arguments:
    cliente -- cliente de la tienda
    """
    self.clientes_fieles.append(cliente)

  def cliente_fiel(self, cliente):
    """Verifica si el cliente es fiel y retorna un booleano"""
    return cliente in self.clientes_fieles

  def producto_registrado(self, producto):
    """Verifica si el producto existe en la tienda y retorna un booleano

    Keyword arguments:
    producto -- producto que se va a buscar
    """
    return producto in self.productos

  def hacer_pedido(self, cliente, productos):
    """Crea el pedido que va a hacer el cliente y retorna el
    identificador del pedido

    Keyword arguments:
    cliente -- cliente que realiza el pedido
    productos -- lista de productos que se van pedir
    """
    if self.cliente_fiel(cliente):
      pedido = PedidoDescuento(cliente, self.descuento)
      for producto in productos:
        if self.inventario.producto_registrado(producto):
          pedido.adicionar_producto(producto)
        else:
          print(f'El producto {producto.nombre} no esta en el inventario\n')
    else:
      pedido = Pedido(cliente)
      for producto in productos:
        pedido.adicionar_producto(producto)
    self.historial_pedidos.append(pedido)
    return len(self.historial_pedidos)-1

  def revisar_pedido(self, numero_pedido):
    """Imprime el pedido indicado sino existe imprime un error

    Keyword arguments:
    numero_pedido -- numero que identifica el pedido
    """
    try:
      pedido = self.historial_pedidos[numero_pedido]
      pedido.revisar()
    except Exception as e:
      print(f'Error el numero de pedido {numero_pedido} no existe')


javier = Cliente(1123,'Javier', 'Calle 90 # 31 22')
jon = Cliente(6789,'Jon', 'Calle 14 # 57 A 90')
camisa = Producto(10, 'Camisa', 12000)
pantalon = Producto(11, 'Pantalon', 13000)
correa = Producto(12, 'Correa', 10000)

tienda_ropa = TiendaRopa('Velez')

tienda_ropa.adicionar_producto(camisa)
tienda_ropa.adicionar_producto(pantalon)
tienda_ropa.adicionar_producto(correa)

tienda_ropa.revisar_inventario()

numero_pedido = tienda_ropa.hacer_pedido(javier, [camisa, pantalon])
tienda_ropa.revisar_pedido(numero_pedido)

tienda_ropa.registrar_cliente_fiel(jon)
numero_pedido = tienda_ropa.hacer_pedido(jon, [camisa, pantalon, correa])
tienda_ropa.revisar_pedido(numero_pedido)
```

```python
def calcular_numero_euler(n):
  """Calcula la constante de Napier para n términos.

  Retorna un areglo en el cual se pueden visualizar los valores decimales,
  para cada cálculo de los n términos

  Keyword arguments:
  n -- número de terminos
  """
  return sum(1/np.vectorize(np.math.factorial)(np.arange(n)))

np.vectorize(calcular_numero_euler)(np.arange(1, 11))
```

```python
def calcular_auto_valores(matriz):
  """Calcula los auto valores de una matriz dada

  Se debe ingresar una lista con sublistas de dimension cuadrada (m*m)

  Ej: calcular_auto_valores([[2, 1, 1],[4, 3, 2],[1, 1, 2]])

  Keyword arguments:
  matriz -- lista de dimension cuadrada (m*m)
  """
  autovalores = np.linalg.eigvals(matriz)
  print(f"Los autovalores de la matriz son: {autovalores}")

calcular_auto_valores([[2, 1, 1],[4, 3, 2],[1, 1, 2]])
```
